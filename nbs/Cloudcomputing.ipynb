```python
content_ratings = df_apps.Content_Rating.value_counts()
content_ratings
```

```python
df_apps.info()
```

```python
df_apps.Installs.describe()
```

```python
df_apps[['App','Installs']].groupby("Installs").count()
```

```python
df_apps.Installs = df_apps.Installs.astype(str).str.replace(',', "")
df_apps.Installs = pd.to_numeric(df_apps.Installs)
df_apps[['App', 'Installs']].groupby('Installs').count()
```

```python
df_apps.Price = df_apps.Price.astype(str).str.replace('$', "")
```

```python
df_apps.Price = pd.to_numeric(df_apps.Price)
```

```python
df_apps.sort_values('Price', ascending=False).head(20)
```

```python
df_apps = df_apps[df_apps["Price"] < 250]
df_apps.sort_values('Price', ascending=False).head(5)
```

```python
df_apps["Approx_Revenue"] = df_apps.Price.mul(df_apps.Installs)

df_apps
```

```python
df_apps.sort_values("Approx_Revenue", ascending=False)
```

```python
df_apps.Category.nunique()
```

```python
most_app_category = df_apps.Category.value_counts()[:10]

most_app_category
```

```python
bar_graph = px.bar(x=most_app_category.index, y=most_app_category.values)

bar_graph.show()
```

```python
Installs_category = df_apps.groupby('Category').agg({'Installs': pd.Series.sum})
Installs_category.sort_values('Installs', ascending=False, inplace=True)
Installs_category[:10]
```

```python
bar_graph2 = px.bar(
    x = Installs_category.Installs,
    y = Installs_category.index,
    title = "Number of installations in each category",
  )

bar_graph2.update_layout(
    xaxis_title= "Number of Downloads",
    yaxis_title= "Category"
)

bar_graph2.show()
```

```python
app_in_category = df_apps.groupby('Category').agg({'App': pd.Series.count})
```

```python
merged_df = pd.merge(app_in_category, Installs_category, on ="Category", how="inner")
```

```python
merged_df.shape
```

```python
merged_df.sort_values("Installs", ascending=False)[:10]
```

```python
scatter_plot = px.scatter(
    data_frame=merged_df,
    x='App',
    y="Installs",
    title = "Installs in Each Category",
    color = "Installs",
    hover_name=merged_df.index
)

scatter_plot.update_layout(
    xaxis_title="Number of Apps",
    yaxis_title="Installs",
    yaxis=dict(type='log')
)

scatter_plot.show()
```

```python
df_apps.Type.value_counts()
```

```python
df_free_vs_paid = df_apps.groupby(["Category", "Type"], as_index=False).agg({'App': pd.Series.count})
df_free_vs_paid.sort_values("App", ascending=False)[:10]
```

```python
bar_graph4 = px.bar(
    df_free_vs_paid,
    x='Category',
    y ='App',
    title='Free vs Paid Apps by Category',
    color='Type',
    barmode='group',
  )
 
bar_graph4.update_layout(
    xaxis_title='Category',
    yaxis_title='Number of Apps',
    xaxis={'categoryorder':'total descending'},
    yaxis=dict(type='log')
)

bar_graph4.show()
```

```python
boxplot1 = px.box(
    df_apps,
    y="Installs",
    x="Type",
    color = "Type",
    notched = True,
    points="all",
    title="Difference in Downloads"
)

boxplot1.update_layout(yaxis=dict(type='log'))

boxplot1.show()
```

```python
df_paid_apps = df_apps[df_apps['Type'] == 'Paid']
```

```python
boxplot2 = px.box(
    df_paid_apps, 
    x='Category', 
    y='Approx_Revenue',
    title='Revenue of paid Apps'
)

boxplot2.update_layout(
    xaxis_title='Category',
    yaxis_title='Paid App Revenue',
    xaxis={'categoryorder':'min ascending'},
    yaxis=dict(type='log')
)
 
 
boxplot2.show()
```

```python
print(df_paid_apps.Approx_Revenue.median())
print(df_paid_apps.Price.median())
```

```python
boxplot3 = px.box(
    df_paid_apps,
    x='Category',
    y="Price",
    title='Price per Category'
)
 
boxplot3.update_layout(
    xaxis_title='Category',
    yaxis_title='App Price',
    xaxis={'categoryorder':'max descending'},
    yaxis=dict(type='log')
)
 
boxplot3.show()
```
